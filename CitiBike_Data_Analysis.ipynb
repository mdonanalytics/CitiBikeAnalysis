```python
import pandas as pd
```

```python

# 1. Create a DataFrame for the Citibike data.
file_to_read = pd.read_csv('/content/201306-citibike-tripdata.csv')
citi_df = file_to_read

```

```python
# 2 Display dataset insights
citi_df.info() # field types
citi_df.isna().sum() # total number of blank values per field
```

```python
# 3 Display the current DataFrame
citi_df.head() # tripduration is displayed in seconds
```

```python
# 4. Convert the 'tripduration' column to minutes
citi_df['tripduration'] = citi_df['tripduration'] / 60
citi_df.head()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
citi_df.to_csv('/content/201306-citibike_cleaned-tripdata.csv')
print("CitiBike CSV Exported Successfully")
```
